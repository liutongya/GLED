## Initial positions of Lagrangian particles

The package called Floater (https://github.com/ocean-transport/floater) is employed to generate the initial float file. 

```python
from floater import generators
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
import xarray as xr
from xmitgcm import open_mdsdataset
```

```python
path = '/data/home/liutongya/RCLV/velocity/run_offline/run/'
ds = open_mdsdataset(path, prefix={'Diag_stat'}, iters=1).chunk()
ds
```

```python
lon = ds.XC.values
lat = ds.YC.values
land_mask = ds.hFacC.values==1

# The user can make their own land mask
```

```python
model_grid = {'lon': lon, 'lat': lat, 'land_mask': land_mask}

fs = generators.FloatSet(xlim=(0, 360), ylim=(-80.0, 80.0), dx=0.03125, dy =0.03125, zvect=np.array(-5), model_grid=model_grid)
fs.to_mitgcm_format('/data/home/liutongya/RCLV/notebooks/files/flt_ini_pos_hex.epac.32deg.bin', mesh='rect', read_binary_prec=32, iup=0)
fs.to_pickle('/data/home/liutongya/RCLV/notebooks/files/floatset.pkl')
```

```python
fig = plt.figure(figsize=(12, 6))
plt.pcolormesh(fs.model_grid['land_mask'], cmap='seismic_r')
plt.colorbar()
```
