## Global LAVD field

```python
import numpy as np
import xarray as xr
from floater import rclv
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from matplotlib import lines
import cmocean
```

```python
import cartopy.crs as ccrs
import cartopy.feature as cft
import matplotlib.ticker as mticker
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
```

```python
ds = xr.open_mfdataset('/data/home/liutongya/RCLV/lavd_netcdf/2016-10-01/float_trajectories_netcdf/*.nc')
ds
```

```python
n_time = 9

xx0 = 95; xx1 = 110;
yy0 = -40; yy1 = -25;

lx0 = ds.sel(x0=slice(xx0,xx1), y0=slice(yy0,yy1)).x[0, :, :]
lx1 = ds.sel(x0=slice(xx0,xx1), y0=slice(yy0,yy1)).x[n_time, :, :]

ly0 = ds.sel(x0=slice(xx0,xx1), y0=slice(yy0,yy1)).y[0, :, :]
ly1 = ds.sel(x0=slice(xx0,xx1), y0=slice(yy0,yy1)).y[n_time, :, :]

lxt = (lx0.values, lx1.values)
lyt = (ly0.values, ly1.values)

lavd1 = ds.sel(x0=slice(xx0,xx1), y0=slice(yy0,yy1)).lavd[n_time, :, :].to_masked_array().filled(0)

# ci_th -1 cd 0.1
kwargs = dict(CI_th=-1, CI_tol=0.1, min_distance=13, min_limit_diff=1e-8,
              min_area=169, max_footprint=100000, use_threadpool=True, progress=True, convex_def=0.1, 
              max_width=100, min_peak=1e-6)

contours1 = list(rclv.find_convex_contours(lavd1, lxt, lyt, **kwargs))

lon1 = ds.sel(x0=slice(xx0,xx1)).x0
lat1 = ds.sel(y0=slice(yy0,yy1)).y0

all_contours = [r[1] for r in contours1]
labels1 = rclv.label_points_in_contours(lavd1.shape, all_contours)
labels1[labels1>0] = 1
labels1 = np.ma.masked_array(labels1, mask=labels1==0)

pos_x0_1 = lx0.values * labels1
pos_y0_1 = ly0.values * labels1

pos_x1_1 = lx1.values * labels1
pos_y1_1 = ly1.values * labels1
```

```python
n_time = 3

xx0 = 290; xx1 = 300;
yy0 = 30; yy1 = 40;

lx0 = ds.sel(x0=slice(xx0,xx1), y0=slice(yy0,yy1)).x[0, :, :]
lx1 = ds.sel(x0=slice(xx0,xx1), y0=slice(yy0,yy1)).x[n_time, :, :]

ly0 = ds.sel(x0=slice(xx0,xx1), y0=slice(yy0,yy1)).y[0, :, :]
ly1 = ds.sel(x0=slice(xx0,xx1), y0=slice(yy0,yy1)).y[n_time, :, :]

lxt = (lx0.values, lx1.values)
lyt = (ly0.values, ly1.values)

lavd2 = ds.sel(x0=slice(xx0,xx1), y0=slice(yy0,yy1)).lavd[n_time, :, :].to_masked_array().filled(0)

# ci_th -1 cd 0.1
kwargs = dict(CI_th=-1, CI_tol=0.1, min_distance=13, min_limit_diff=1e-8,
              min_area=169, max_footprint=100000, use_threadpool=True, progress=True, convex_def=0.1, 
              max_width=100, min_peak=1e-6)

contours2 = list(rclv.find_convex_contours(lavd2, lxt, lyt, **kwargs))

lon2 = ds.sel(x0=slice(xx0,xx1)).x0
lat2 = ds.sel(y0=slice(yy0,yy1)).y0

all_contours = [r[1] for r in contours2]
labels2 = rclv.label_points_in_contours(lavd2.shape, all_contours)
labels2[labels2>0] = 1
labels2 = np.ma.masked_array(labels2, mask=labels2==0)

pos_x0_2 = lx0.values * labels2
pos_y0_2 = ly0.values * labels2

pos_x1_2 = lx1.values * labels2
pos_y1_2 = ly1.values * labels2
```
